```python
# üõ†Ô∏è Install required package
!pip install -q --upgrade bar_chart_race

# üìö Import libraries
import pandas as pd
from google.colab import files
import bar_chart_race as bcr

# üìÇ Upload CSV file
print("üìÇ Please upload your CSV file...")
uploaded = files.upload()
filename = next(iter(uploaded))

# üßº Read and clean the data
df = pd.read_csv(filename, index_col=0)
df.index.name = 'Country'
df = df.dropna(how='all', axis=1)
df = df.apply(pd.to_numeric, errors='coerce')
df = df.dropna(how='all')

# Transpose and convert index to datetime
df_T = df.transpose()
df_T.index.name = 'Year'
df_T.index = pd.to_datetime(df_T.index, format='%Y')

output_filename = 'bar_race.mp4'

# Create bar chart race animation (no direct way to hide x-axis)
bcr.bar_chart_race(
    df=df_T,
    filename=output_filename,
    orientation='h',
    sort='desc',
    n_bars=7,
    fixed_order=False,
    fixed_max=True,
    steps_per_period=25,
    interpolate_period=True,
    period_length=500,
    title='Top 7 Populated Countries Over Time',
    bar_size=0.95,
    scale='linear',
    cmap='dark12',
    period_label={'x': .99, 'y': .15, 'ha': 'right', 'va': 'center'},
    period_fmt='%Y',
    bar_kwargs={'alpha': 0.8, 'ec': 'black'},
    filter_column_colors=True
)

print(f"‚úÖ Bar chart race saved as: {output_filename}")
```
